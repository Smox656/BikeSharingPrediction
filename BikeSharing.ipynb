```python
periods = {"day" : [8, 9, 10, 11, 12, 13, 14, 16, 17, 18],
           "evening" : [19, 20, 21, 22, 23],
           "night" : [0, 1, 2, 3, 4, 5, 6, 7]}
```

```python
def which_period(line):
  for period, hours in periods.items():
    if line['hr'] in hours:
      p = period
      if p == "day":
        return 0
      elif p == "evening":
        return 1
      else:
        return 2
```

```python
which_period(df.iloc[0])
```

```python
df['period'] = df.apply(which_period, axis=1)
```
